### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

```python
# Write your code here
print(df.info())
```

#### 3. Identifying and Removing Inconsistencies

<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# Check unique values to find inconsistencies
print(df['Country'].unique())
# Or count them to see rare/weird entries
print(df['Country'].value_counts().tail(20))
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
#Xử lý khoảng trắng, NaN, format chữ
df['Country'] = (
    df['Country']
    .astype(str)
    .str.strip()
    .replace('nan', pd.NA)
)

df['Country'] = df['Country'].fillna('Unknown')
```

```python
# Chuẩn hóa các tên quốc gia “quá dài / không thực tế”
country_mapping = {
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'Iran, Islamic Republic of...': 'Iran',
    'Korea, Republic of': 'South Korea',
    'Korea, Democratic People\'s Republic of': 'North Korea',
    'Russian Federation': 'Russia',
    'Viet Nam': 'Vietnam',
    'Taiwan, Province of China': 'Taiwan',
    'Hong Kong (S.A.R.)': 'Hong Kong',
    'Micronesia, Federated States of': 'Micronesia',
    'Bolivia, Plurinational State of': 'Bolivia',
    'Tanzania, United Republic of': 'Tanzania',
    'Moldova, Republic of': 'Moldova'
}

df['Country'] = df['Country'].replace(country_mapping)
```

```python
## Write your code here
#Gộp country hiếm
country_counts = df['Country'].value_counts()
threshold = 50  # bạn có thể đổi: 20 / 100 tùy bài toán
rare_countries = country_counts[country_counts < threshold].index

df['Country'] = df['Country'].replace(rare_countries, 'Other')
```

```python
## Write your code here
print(df['Country'].value_counts().head(20))
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df = pd.get_dummies(df, columns=['Employment'], prefix='Employment')
print(df.head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_counts = df.isnull().sum()
print("Top missing columns:\n", missing_counts.sort_values(ascending=False).head())
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
median_comp = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_comp)
print("Missing ConvertedCompYearly after fill:", df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
remote_mode = df['RemoteWork'].mode()[0]
df['RemoteWork']= df['RemoteWork'].fillna(remote_mode)
print("Missing RemoteWork after fill:", df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
min_val = df['ConvertedCompYearly'].min()
max_val = df['ConvertedCompYearly'].max()
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - min_val) / (max_val - min_val)
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np
# Adding 1 to avoid log(0) duplicate error if any zero exists, though salary usually > 0
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df['YearsCodePro'].unique()
```

```python
def classify_experience(years):
    if pd.isna(years):
        return 'Unknown'
    if years == 'Less than 1 year':
        return 'Junior'
    if years == 'More than 50 years':
        return 'Senior'
    
    try:
        y = float(years)
        if y <= 5:
            return 'Junior'
        elif y <= 10:
            return 'Mid-level'
        else:
            return 'Senior'
    except:
        return 'Unknown'

df['ExperienceLevel'] = df['YearsCodePro'].apply(classify_experience)

# Check distribution
print(df['ExperienceLevel'].value_counts())
```
